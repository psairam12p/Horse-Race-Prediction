# â—‹	Scatter plots, histograms, and box plots to visualize data distribution

```python
# detecting the skewed columns using plot
def plot(df,column):
  #distplot
  plt.figure(figsize=(15,4))
  plt.subplot(1,3,1)
  sns.distplot(df[column])
  plt.title("distplot for"+" "+column)

  #histogram plot

  plt.subplot(1,3,2)
  sns.histplot(df, x= column, kde= True, bins=30,color="salmon")
  plt.title("histogram plot for"+" "+column)

  #boxplot

  plt.subplot(1,3,3)

  sns.boxplot(df, x=column)
  plt.title("Box plot for"+" "+column)
```

```python
for i in df_resampled.columns:
    plot(df_resampled,i)
```

```python
# outlier columns are 
# 'age','decimalPrice','isFav','position','positionL','dist','weightSt','RPR','TR','OR','runners','weight'
```

```python
def outlier(df_resampled,column):
  q1= df_resampled[column].quantile(0.25)
  q3= df_resampled[column].quantile(0.75)

  iqr= q3-q1

  upper_threshold= q3 + (1.5*iqr)
  lower_threshold= q1 - (1.5*iqr)

  df_resampled[column]= df_resampled[column].clip(lower_threshold, upper_threshold)
```

```python
outlier_columns=['age','decimalPrice','isFav','position','positionL','dist','weightSt','RPR','TR','OR','runners','weight']
for j in outlier_columns:
    outlier(df_resampled,j)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,auc,roc_curve,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=df_resampled.drop(['res_place','res_win'],axis=1)
y=df_resampled['res_win']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
models=[LogisticRegression(),DecisionTreeClassifier(),RandomForestClassifier(),KNeighborsClassifier()]
```

```python

for model in models:
    model.fit(x_train,y_train)
    train_pred=model.predict(x_train)
    test_pred=model.predict(x_test)

    print(f"*************{type(model).__name__}************")
    print(f"Train Accuracy: {accuracy_score(y_train,train_pred)}")
    print(f"Train Precision: {precision_score(y_train,train_pred)}")
    print(f"Train Recall: {recall_score(y_train,train_pred)}")
    print(f"Train Precision: {f1_score(y_train,train_pred)}")
    print("...............................................")
    print(f"Test Accuracy: {accuracy_score(y_test,test_pred)}")
    print(f"Test Precision: {precision_score(y_test,test_pred)}")
    print(f"Test Recall: {recall_score(y_test,test_pred)}")
    print(f"Test Precision: {f1_score(y_test,test_pred)}")
```

```python
# Get the high accuracy using Hyperparameter Tuning method with using of the GridsearchCV

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.2, random_state=42)

parameters= {"max_depth": [5, 10],
             "min_samples_split": [2,5],
             "min_samples_leaf": [1,2],
             "max_features": ['sqrt', 'log2']
             }

gridsearch= GridSearchCV(estimator= DecisionTreeClassifier(), param_grid= parameters, cv= 3, n_jobs= -1,verbose=1)
gridsearch.fit(x_train,y_train)
```

```python
# Passing the best Hypertuning paramers in the ,
# RandomForest algorithm and check the accuracy for training and testing
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state= 42)

model= DecisionTreeClassifier(max_depth=10, max_features= 'sqrt', min_samples_leaf=2, min_samples_split=5).fit(x_train,y_train)

y_pred_train= model.predict(x_train)
y_pred_test= model.predict(x_test)

#checking the accuracy_score for train and test

accuracy_train= accuracy_score(y_train, y_pred_train)
accuracy_test= accuracy_score(y_test, y_pred_test)

print("Accuracy score for Train and Test")
print("----------------------------------")
print("Accuracy_Train: ",accuracy_train)
print("Accuracy_Test: ",accuracy_test)
print("  ")
#confution matrics and the classification report for test

print("Confution_matrix for Test")
print("--------------------------")
print(confusion_matrix(y_true= y_test, y_pred= y_pred_test))
print(" ")
print("Classification_report for Test")
print("-------------------------------")
print(classification_report(y_true= y_test, y_pred= y_pred_test))

```

```python
# Receiver Operating Characteristic (ROC) Curve and Area Under the Curve (AUC)

FP,TP,threshold= roc_curve(y_true= y_test, y_score=y_pred_test)
print(threshold)
print(FP)
print(TP)
print(" ")
auc_curve= auc(x=FP,y=TP)
print("auc_curve:",auc_curve)
```

```python
# create a plot for roc and auc curve
roc_point= {"ROC Curve (area)":round(auc_curve,2)}
plt.plot(FP,TP,label= roc_point)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.1])
plt.xlabel("False Positive")
plt.ylabel("True Positive")
plt.plot([0,1],[0,1],"k--")
plt.legend(loc= "lower right")
plt.show()
```
